```python
if df_1['similarities'][0] > df_2['similarities'][0]:
  relevent_section=df_1
else:
  relavent_section=df_2
if relavent_section['similarities'][0] > df_3['similarities'][0]:
  relevant_section=relavent_section
else:
  relevant_section=df_3
relevant_section
```

```python
def generate_answer(question, relevant_section):
    prompt = f"Question: {question}\nContext: {relevant_section['text']}\nAnswer:"
    response = openai.Completion.create(
        engine='text-davinci-003',
        prompt=prompt,
        max_tokens=100,
        temperature=0.7,
        n=1,
        logprobs=0,
        stop=None,
        #temperature=0.7
    )
    answer = response.choices[0]['text'].strip()
    return answer
```

```python
generate_answer(question,relevant_section)
```
